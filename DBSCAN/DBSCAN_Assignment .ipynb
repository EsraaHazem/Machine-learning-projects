```python
#!pip install pandas plotnine
# use any plotting library seaborn matplotlib,..........
from sklearn.neighbors import NearestNeighbors

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
import pandas as pd
from sklearn.preprocessing import StandardScaler
from plotnine import *
import sklearn.datasets
from sklearn.cluster import DBSCAN ,AgglomerativeClustering,KMeans
import seaborn as sns
from sklearn.metrics import silhouette_score
```

```python
n_samples = 1500
dataset1 =sklearn.datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
dataset1=pd.DataFrame(dataset1[0])
dataset2 = sklearn.datasets.make_moons(n_samples=n_samples, noise=.05)
dataset2=pd.DataFrame(dataset2[0])

dataset3 = sklearn.datasets.make_blobs(n_samples=n_samples, random_state=8)
dataset3=pd.DataFrame(dataset3[0])

dataset4= np.random.rand(n_samples, 2), None
dataset4=pd.DataFrame(dataset4[0])

dataset5= pd.read_csv("dataset5.csv")
dataset6= pd.read_csv("dataset6.csv")
dataset7= pd.read_csv("dataset7.csv")
dataset8= pd.read_csv("dataset8.csv")
dataset9= pd.read_csv("dataset9.csv")

```

1) plot all dataset
2) use standard scalar
3) use K-means, agglomerative, dbscan
4) plot dataset after clustering
5) Evaluate inertia,silhouette score,..........

```python
mins = 3
nn = NearestNeighbors(n_neighbors=4)

nn.fit(dataset9[["x", "y"]])

distances, neighbors = nn.kneighbors(dataset9[["x", "y"]])

distances = np.sort(distances[:, mins], axis = 0)
X=list(range(0,len(distances)))
plt.plot(X,distances)
```

```python
# plotting the data
scaler1 = StandardScaler()
dataset1[['x','y']] = scaler1.fit_transform(dataset1)
ggplot(dataset1, aes(x = "x", y = "y")) + geom_point() 
```

```python
    fig,ax = plt.subplots(1,3,figsize=(15,5))

    dbscan = DBSCAN(eps=0.2,min_samples=5)
    dbscan.fit(dataset1[['x','y']])
    dataset1['clusters_db'] = pd.Categorical(dbscan.labels_)
    sns.scatterplot(x='x',y='y',hue='clusters_db',data = dataset1,ax=ax[0])

    agglom_ward = AgglomerativeClustering(n_clusters=2,linkage='ward')
    agglom_ward.fit(dataset1[['x','y']])
    dataset1['clusters_agg'] = pd.Categorical(agglom_ward.labels_)
    sns.scatterplot(x='x',y='y',hue='clusters_agg',data = dataset1,ax=ax[1])

    km = KMeans(n_clusters=2)
    km.fit(dataset1[['x','y']])
    dataset1['clusters_km'] = pd.Categorical(km.labels_)
    sns.scatterplot(x='x',y='y',hue='clusters_km',data = dataset1,ax=ax[2])

    ax[0].title.set_text('DBSCAN')
    ax[1].title.set_text('Agglomerative')
    ax[2].title.set_text('kmeans')
```

```python
from sklearn.metrics import silhouette_score
print('siloete score - DBSCAN : ',silhouette_score(dataset1[['x','y']],dataset1['clusters_db']))
print('siloete score - KMeans : ',silhouette_score(dataset1[['x','y']],dataset1['clusters_km']))
print('siloete score - Agglomerative-ward : ',silhouette_score(dataset1[['x','y']],dataset1['clusters_agg']))
```
